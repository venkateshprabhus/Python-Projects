```python
Trip_data['started_at'] = pd.to_datetime(Trip_data['started_at'])
Trip_data['ended_at'] = pd.to_datetime(Trip_data['ended_at'])
```

```python
Trip_data['hour'] = Trip_data['started_at'].dt.hour
Trip_data = Trip_data.astype({'hour' : 'int8'})
```

```python
Trip_data['year'] = Trip_data['started_at'].dt.year
Trip_data = Trip_data.astype({'year' : 'int16'})
```

```python
Trip_data['dayofweek'] = Trip_data['started_at'].dt.day_name()
Trip_data['month'] = Trip_data['started_at'].dt.month_name()
```

```python
Trip_data['ride_length_min'] = (Trip_data['ended_at'] - Trip_data['started_at'])/dt.timedelta(minutes = 1)
Trip_data['ride_length_min'] = Trip_data['ride_length_min'].astype('int32')
```

```python
Trip_data[Trip_data['ride_length_min'] <= 0].count()
```

```python
Trip_data = Trip_data[Trip_data['ride_length_min'] >= 1].reset_index()
Trip_data = Trip_data.drop(columns=['index'])
```

```python
Trip_data['ride_length_hour'] = round(Trip_data['ride_length_min']/60, 2)
```
